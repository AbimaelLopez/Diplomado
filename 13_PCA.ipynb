## **1. Importar bibliotecas**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.datasets import fetch_olivetti_faces
```

## **2. Uso de PCA en un conjunto de datos sintético**

### **2.1 Creación y graficación de un conjunto de datos sintético**

```python
# Create a synthetic dataset with three blobs
X, y = make_blobs(n_samples=100, n_features=2, centers=3, random_state=42)

# Visualize the original data
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', edgecolors='k')
plt.title('Original Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

### **2.2 Aplicación de PCA**

```python

# Apply PCA to reduce the dimensionality to 1
pca = PCA(n_components=1)
X_pca = pca.fit_transform(X)

# Visualize the data after PCA
plt.figure(figsize=(8, 6))
plt.scatter(X_pca, np.zeros_like(X_pca), c=y, cmap='viridis', edgecolors='k')
plt.title('Data After PCA (Reduced to 1 Dimension)')
plt.xlabel('Principal Component 1')
plt.yticks([])  # No y-axis ticks since we only have one dimension
plt.show()

# Variance explained by each principal component
explained_variance_ratio = pca.explained_variance_ratio_
print(f"Variance explained by Principal Component 1: {explained_variance_ratio[0]:.2f}")
```

## **3. Uso de PCA sobre el conjunto de datos Iris**


*   Se carga el conjunto de datos Iris
*   Se separa en la matriz X y y (solo para comparación)
*   Se aplica PCA
*   Se visualizan los datos después de aplicar PCA
*   Se indica que tanta varianza representan los componentes
*   Con propósitos de comparación, se grafican todas las combinaciones de características y el resultado de PCA


```python
# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names

# Apply PCA to reduce the dimensionality to 2
n_components=2

pca = PCA(n_components=n_components)
X_pca = pca.fit_transform(X)

# Visualize the data after PCA
plt.figure(figsize=(6, 6))

for i, target_name in enumerate(target_names):
    plt.scatter(X_pca[y == i, 0], X_pca[y == i, 1], label=target_name)

plt.title('Iris Dataset After PCA (2D)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()

# Variance explained by each principal component
explained_variance_ratio = pca.explained_variance_ratio_
print(f"Variance explained by Principal Component 1: {explained_variance_ratio[0]:.2f}")
print(f"Variance explained by Principal Component 2: {explained_variance_ratio[1]:.2f}")
```

### **4. Uso de PCA para reducir dimensionalidad de conjunto de datos de rostros**


*   Lectura de datos
*   Separación en X y y
*   Se grafican los rostros originales
*   Se usa PCA para reducir la dimensionalidad
*   Se grafican los datos en la dimensión reducida
*   Se aplica una transformación inversa para reconstruir los datos
*   Se grafican los rostros reconstruidos
*   Se sugiere probar con más componentes para observar como afecta en la reconstruicción. NOTA: La grafica que despliega los puntos con dimensión reducida solo considera las dos primeras coordenadas.



```python
# Load the Olivetti Faces dataset
faces_data = fetch_olivetti_faces(shuffle=True, random_state=42)
X_faces = faces_data.data
y_faces = faces_data.target

# Display some original facial images
plt.figure(figsize=(12, 6))
for i in range(6):
    plt.subplot(2, 6, i + 1)
    plt.imshow(X_faces[i].reshape(64, 64), cmap='gray')
    plt.title(f"Person {y_faces[i]}")
    plt.axis('off')

# Apply PCA to reduce the dimensionality to 2
n_components=2
pca_faces = PCA(n_components=n_components)
X_faces_pca = pca_faces.fit_transform(X_faces)

# Visualize the data after PCA
plt.subplot(2, 2, 3)
for i in np.unique(y_faces):
    plt.scatter(X_faces_pca[y_faces == i, 0], X_faces_pca[y_faces == i, 1], label=f'Person {i}')

plt.title('Olivetti Faces Dataset After PCA (2D)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()


```

```python
# Display some original facial images
plt.figure(figsize=(14, 7))
for i in range(6):
    plt.subplot(2, 6, i + 1)
    plt.imshow(X_faces[i].reshape(64, 64), cmap='gray')
    plt.title(f"Person {y_faces[i]}")
    plt.axis('off')

# Display some reconstructed facial images after PCA
X_faces_reconstructed = pca_faces.inverse_transform(X_faces_pca)
for i in range(6):
    plt.subplot(2, 6, i + 7)
    plt.imshow(X_faces_reconstructed[i].reshape(64, 64), cmap='gray')
    plt.title(f"Person {y_faces[i]} (Reconstructed)")
    plt.axis('off')

plt.suptitle('Original Facial Images and PCA Results')
plt.tight_layout()
plt.show()

# Variance explained by each principal component
explained_variance_ratio_faces = pca_faces.explained_variance_ratio_
print(f"Variance explained by Principal Component 1: {explained_variance_ratio_faces[0]:.2f}")
print(f"Variance explained by Principal Component 2: {explained_variance_ratio_faces[1]:.2f}")
```
